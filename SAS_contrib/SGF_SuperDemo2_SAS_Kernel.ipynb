# This is a SAS kernel; type in SAS code ...

## If there is List output, SAS_Kernel displays that by default (ODS HTML5 Output)

```sas
proc print data=sashelp.cars(obs=5);run;
```

## If there is no List output, it displays the SASLOG

```sas
data saskernel1; set sashelp.cars;run;
```

## Errors will get you the log too

```sas
proc print data=saskernel1(flobs=5);run;
```

## Lets run the Heatmap code from saspy

```sas
proc sgplot data=sashelp.cars ;
    heatmap x=msrp y=horsepower;;
run;
```

## Let's assign our libref and see this is a different SAS Session than any of them in the saspy Notebook

```sas
libname libk 'c:\public\saspy_demo\libk';
```

```sas
        data _null_; retain libref; retain cobs 1; 
           set sashelp.vlibnam end=last;
           if cobs EQ 1 then
              put "LIBREFSSTART";
           cobs = 2;
           if libref NE libname then
              put "LIBREF=" libname;
           libref = libname;
           if last then
              put "LIBREFSEND";
        run;
```

```sas
proc sql;select distinct(libname) from sashelp.vlibnam; quit;
```
